```python
import pandas as pd
```

```python
df=pd.DataFrame({'Age' : [23,35,27,19, 42],
               'Gender':["M","F","M","M", "F"],
                'Education' : ["Bachelor", "High School", "Masters","Bachelor", "Doctorate"]})
```

```python
df
```

```python
bins= [0,18,25,35,60]
labels =["<18","18-25","25-35","35+"]
df['Age_Groups'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df
```

```python
gender_dict={'M':0,'F':1}
df['Gender_Code'] =df['Gender'].map(gender_dict)
```

```python
df
```

```python
edu_dict ={'High School': 1, "Bachelor": 2, "Masters":3, "Doctorate": 4}
df['Education_level'] = df['Education'].map(edu_dict)
```
